```python
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Exemplo de dados para modelagem preditiva
data_model = {
    'latitude': [23.5, 24.1, 22.8, 23.9, 24.3],
    'longitude': [45.2, 46.1, 44.8, 45.9, 46.3],
    'area_hectares': [100, 150, 120, 130, 140],
    'terrain_type': [1, 2, 1, 3, 2],  # Urbano=1, Rural=2, Montanhoso=3
    'weather_conditions': [1, 2, 1, 3, 2],  # Bom=1, Regular=2, Ruim=3
    'drone_usage_efficiency': [80, 85, 78, 82, 84]  # Eficiência em %
}

# Criar DataFrame para modelagem
df_model = pd.DataFrame(data_model)

# Definir variáveis independentes (X) e variável dependente (y)
X = df_model.drop('drone_usage_efficiency', axis=1)
y = df_model['drone_usage_efficiency']

# Dividir dados em treino e teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.4, random_state=42)

# Criar modelo de regressão linear
model = LinearRegression()
model.fit(X_train, y_train)

# Fazer previsões
y_pred = model.predict(X_test)

# Avaliar modelo
print('Mean Squared Error:', mean_squared_error(y_test, y_pred))
print('R2 Score:', r2_score(y_test, y_pred))

# Exibir coeficientes do modelo
print('Coeficientes:', model.coef_)
print('Intercepto:', model.intercept_)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import pi

# Carregar dados do arquivo Excel
file_path = 'Survey_ DRONES - 13_01_25 1.xlsx'
df = pd.read_excel(file_path, engine='openpyxl')

# Exibir as primeiras linhas do DataFrame
print(df.head())

# Exemplo fictício de dados para gráficos de radar
data_radar = {
    'latitude': [23.5, 24.1, 22.8, 23.9, 24.3],
    'longitude': [45.2, 46.1, 44.8, 45.9, 46.3],
    'area_hectares': [100, 150, 120, 130, 140],
    'terrain_type': [1, 2, 1, 3, 2],  # Urbano=1, Rural=2, Montanhoso=3
    'weather_conditions': [1, 2, 1, 3, 2],  # Bom=1, Regular=2, Ruim=3
    'drone_usage_efficiency': [80, 85, 78, 82, 84]  # Eficiência em %
}

# Criar DataFrame para gráficos de radar
df_radar = pd.DataFrame(data_radar)

# Função para criar gráfico de radar
def create_radar_chart(df_radar):
    categories = list(df_radar.columns)
    N = len(categories)
    
    # Valores médios para cada categoria
    values = df_radar.mean().tolist()
    values += values[:1]  # Repetir o primeiro valor para fechar o círculo
    
    # Ângulos para cada categoria
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]  # Repetir o primeiro ângulo para fechar o círculo
    
    # Inicializar o gráfico de radar
    ax = plt.subplot(111, polar=True)
    
    # Desenhar um eixo para cada categoria e adicionar rótulos
    plt.xticks(angles[:-1], categories)
    
    # Desenhar os dados do gráfico de radar
    ax.plot(angles, values, linewidth=2, linestyle='solid')
    
    # Preencher a área do gráfico de radar
    ax.fill(angles, values, 'b', alpha=0.1)
    
    plt.title('Gráfico de Radar - Eficiência do Uso de Drones')
    plt.show()

# Criar gráfico de radar usando os dados fictícios
create_radar_chart(df_radar)
```
