# Gráfico de Radar - Pontuação por Categoria

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from math import pi

# Carregar dados do arquivo Excel
file_path = 'Survey_ DRONES -_01_25 Diretor DTIC Coronel Shoichi 5.xlsx'
df = pd.read_excel(file_path, engine='openpyxl')

# Exibir as primeiras linhas do DataFrame
print(df.head())

# Remover linhas desnecessárias e renomear colunas
df_cleaned = df.drop([0, 1, 2, 3])
df_cleaned.columns = ['Objetivo do Site Survey', 'Ambiente', 'Virtual (SIM / NÃO)', 'Versão / Detalhamento', 'Obs']

# Exibir as primeiras linhas do DataFrame limpo
print(df_cleaned.head())

# Salvar o DataFrame limpo em um novo arquivo Excel
df_cleaned.to_excel('Survey_ DRONES_cleaned.xlsx', index=False)

# Análise e Gráficos

# Gráfico de Barras - Pontuação por Categoria
categories = ['Uso dos Drones - Segurança Pública', 'Uso dos Drones - Aplicações militares', 'Uso dos Drones - Outros usos',
              'Aspectos Éticos', 'Imagem', 'Segurança de voo',
              'Regulamentação - Nacional/Internacional', 'Políicas de bkp', 'Tecnologia - Integração com IA',
              'Tecnologia - Transmissão (RF/4G)', 'Redundância', 'Treinamento',
              'Desafios/Expectativas - Volume de dados', 'Desafios/Expectativas - Migração para nuvem']

scores = [8, 7, 6, 5, 5, 8, 7, 6, 7, 6, 5, 9, 7, 8]

plt.figure(figsize=(10, 6))
plt.barh(categories, scores, color='skyblue')
plt.xlabel('Pontuação')
plt.title('Pontuação por Categoria')
plt.show()

# Gráfico de Radar - Pontuação por Categoria
def create_radar_chart(categories, scores):
    N = len(categories)
    
    # Valores médios para cada categoria
    values = scores
    values += values[:1]  # Repetir o primeiro valor para fechar o círculo
    
    # Ângulos para cada categoria
    angles = [n / float(N) * 2 * pi for n in range(N)]
    angles += angles[:1]  # Repetir o primeiro ângulo para fechar o círculo
    
    # Inicializar o gráfico de radar
    ax = plt.subplot(111, polar=True)
    
    # Desenhar um eixo para cada categoria e adicionar rótulos
    plt.xticks(angles[:-1], categories)
    
    # Desenhar os dados do gráfico de radar
    ax.plot(angles, values, linewidth=2, linestyle='solid')
    
    # Preencher a área do gráfico de radar
    ax.fill(angles, values, 'b', alpha=0.1)
    
    plt.title('Gráfico de Radar - Pontuação por Categoria')
    plt.show()

# Criar gráfico de radar usando os dados fictícios
create_radar_chart(categories, scores)
```
